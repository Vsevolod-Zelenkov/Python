```python
# 3.1 Перечислите все столбцы набора данных и изучите их типы.
    
df.info()
```

```python
    #  #   Column             Non-Null Count  Dtype  
    # ---  ------             --------------  -----  
    #  6   Worldwide Gross    77 non-null     object    - уберем '$', и переведем в числовые данные.
    #  7   Year               77 non-null     int64     - переведем в категориальные данные.

import warnings
warnings.filterwarnings('ignore')

df['Worldwide Gross'] = df['Worldwide Gross'].str.replace('$','').astype(float)
df['Year'] = df['Year'].astype(object)
df.info()
```

```python
    # Видим что это лишние дубликаты в данных. 
        # Уберем их:
    
df = df.drop_duplicates()

    # Убедимся что фильмы у нас не повторяются:

df['Film'].value_counts()
```

```python
    # Проверим жанры.
# (7. Значение «Жанр» временами кажется непоследовательным; попробуйте найти эти несоответствия и исправить их.)

df['Genre'].value_counts()
    
    # Видим непоследовательность:
        # Comedy Comdy comedy
        # Romance Romence romance
```

```python
    # Исправим несоответствия в жанрах:
    
df['Genre'] = df['Genre'].str.title()
df['Genre'] = df['Genre'].str.replace('Comdy','Comedy')
df['Genre'] = df['Genre'].str.replace('Romence','Romance')
df['Genre'].value_counts()
```
