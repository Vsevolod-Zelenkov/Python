```python
# 1. Request an integer as the input. 
    # If it is divisible by 3, output "foo"; 
    # if it is divisible by 5, output "bar"; 
    # if it is divisible by both, print "ham" (not "foo" or "bar").
    
num_1 = int(input("Task 1. Number: "))
if num_1 % 3 == 0 and num_1 % 5 == 0:
    print("ham")
elif num_1 % 3 == 0:
    print("foo")
elif num_1 % 5 == 0:
    print("bar")
else:
    print("not divisible by 3 or 5 without remainder")
```

```python
# 2. Request two numbers as input and print which one is smaller and which one is bigger.

# 2.v1
num_7 = float(input("Task 2.v1 Number_1: "))
num_8 = float(input("Number_2: "))
list_2 = [num_7, num_8]
if num_7 == num_8:
    print("equal")
else:
    print("smaller: " + str(min(list_2)),
          "bigger: " + str(max(list_2)), sep='\n')
```

```python
# 2.v2
num_2 = float(input("Task 2.v2 Number_1: "))
num_3 = float(input("Number_2: "))
if num_2 > num_3: 
    print(str(num_2) + " is bigger", str(num_3) + " is smaller", sep='\n')
elif num_2 < num_3:
    print(str(num_2) + " is smaller", str(num_3) + " is bigger", sep='\n')
else:
    print("equal")
```

```python
# 3. Ask for three numbers as input and print the smallest, the average, and the biggest. 
    # Assume they are all different.

num_4 = float(input("Task 3. Number_1: "))
num_5 = float(input("Number_2: "))
num_6 = float(input("Number_3: "))
list_1 = [num_4, num_5, num_6]
print("min: " + str(min(list_1)),
      "avg: " + str(sum(list_1) / len(list_1)),
      "max: " + str(max(list_1)), sep='\n')
```

```python
# 4. Read this explanation of list slicing - https://www.learnbyexample.org/python-list-slicing/
# 5. Play the game Fizz-Buzz: print all numbers from 1 to 100: 
    # if the number is divisible by 3, print "fizz" instead of the number; 
    # if it's divisible by 5, print "buzz" instead of a number;
    # if it is divisible by both, print "fizz buzz" instead of the number.

start = input("Task 5. Enter 'play' to start: ")
if start == "play":    
    for num in range(1, 101):
        if num % 3 == 0 and num % 5 == 0:
            print("fizz buzz")
        elif num % 3 == 0:
            print("fizz")
        elif num % 5 == 0:
            print("buzz")
        else:
            print(num)
else:
    print("Unknown command. Sorry, see you next time.")
```

```python
# 6. Play the 7-boom game: print all numbers from 1 to 100; 
    # if a number is divisible by 7 or contains the digit 7, print "BOOM" instead of the number.

start = input("Task 6. Enter 'play' to start: ")
if start == "play":
    for num in range(1, 101):
        if num % 7 == 0 or "7" in str(num):
            print("BOOM")
        else:
            print(num)
else:
    print("Unknown command. Sorry, see you next time.")
```
