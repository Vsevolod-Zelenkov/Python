```python
# Используйте модуль времени, чтобы 
    # сравнить производительность «эффективного» метода поиска простых чисел с простой реализацией
            # (без перерывов, тестирования по всем числам и т.д.). 
    # Проверьте несколько диапазонов поиска простых чисел (например, до 100, до 1000 и т. д.)

import time
import math

# Простая реализация__________________________________________________

range_to = int(input("Show prime numbers in range to (including): "))
t0 = time.time()
primes_list = []

for number in range(2, range_to + 1):
    for element in range(2, number):
        if number % element == 0:            
            break
    else:
        primes_list.append(number)

print("",
      "Simple method:",
      "",
      primes_list, sep='\n')
t1 = time.time()
t2 = round(t1-t0, 6)
print("",
      "    - total prime numbers in range: " + str(len(primes_list)),
      "    - time required (sec): " + str(t2),
      "_______________________________________________________________",
      "", sep='\n')



# «Эффективная» реализация____________________________________________

range_to_2 = int(input("To compare with efficient method, please, repeat the input: "))
t3 = time.time()
if range_to_2 != range_to:
    print("",
          "Sorry, you didn't make it, try again.", sep='\n')
else:
    
    def efficient_method(range_to_2):
        """This is more efficient method"""
        
        if range_to_2 >= 2:
            primes_list_2 = [2]
        else:
            primes_list_2 = []

        for number_2 in range(3, range_to_2 + 1, 2):
            is_prime = True
            for element_2 in range(3, int(math.sqrt(number_2)) + 1):
                if number_2 % element_2 == 0:
                    is_prime = False
                    break
            if is_prime:
                primes_list_2.append(number_2)
        return primes_list_2

    print("",
          "Efficient method:",
          "",
          efficient_method(range_to_2), sep='\n')
    t4 = time.time()
    t5 = round(t4-t3, 6)
    if t5 == 0 or t2 <= t5:
        print("",
              "Sorry, try again, maybe with some bigger range.", sep='\n')
    else:
        t6 = round(t2/t5, 1)
        print("",
              "    - total prime numbers in range: " + str(len(efficient_method(range_to_2))),
              "    - time required (sec): " + str(t5),
              "_______________________________________________________________",
              "",
              "Efficient method is {} times faster than simple method (for the input range).".format(t6), sep='\n')
```
