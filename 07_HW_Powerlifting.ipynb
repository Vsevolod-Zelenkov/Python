```python
# Python homework 7

# 1.Прочтите этот https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv
    # набор данных по пауэрлифтингу в pandas DataFrame.
    
import pandas as pd

#df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv')\
                #.drop_duplicates()
    
df = pd.read_csv('ipf_lifts.csv').drop_duplicates()
df
```

```python
df.isna().sum()
```

```python
df['division'] = df['division'].fillna('Without division')
df = df.fillna(0)
df
```

```python
# 2. Найдите рекорды каждого пола и каждого подразделения в каждом из 3 упражнений 
    # (жим лежа, приседания и становая тяга).

# 2.1 Рекорды каждого пола и каждого подразделения в жиме лежа - bench

df.groupby(['division', 'sex'])['best3bench_kg'].max().sort_values(ascending=False)
```

```python
# 2.2 Рекорды каждого пола и каждого подразделения в приседаниях - squat

df.groupby(['division', 'sex'])['best3squat_kg'].max().sort_values(ascending=False)
```

```python
# 2.3 Рекорды каждого пола и каждого подразделения в становой тяге - deadlift

df.groupby(['division', 'sex'])['best3deadlift_kg'].max().sort_values(ascending=False)
```

```python
# 3. Посчитайте количество побед каждого участника, принимая за победу 1-е место. 
    # Сохраните результаты во вторичном DataFrame.
        
        # Принимаем условие как: 
            # Посчитайте Общее количество побед каждого участника за всю его историю соревнований в разных подразделениях.
            
    
df2 = pd.DataFrame(df[df.place == '1'].groupby('name')['place'].count().sort_values(ascending=False))\
                    .rename(columns={'place': 'winnings'})
df2
```

```python
# 4. Используйте DataFrame из упражнения 3, чтобы показать для каждой комбинации подразделения и пола, 
    # участника с наибольшим количеством побед среди участников, когда-либо участвовавших в этом дивизионе.
    
        # Принимаем условие как: 
            # для каждой комбинации подразделения и пола показать участника с наибольшим Общим количеством его побед
            # за всю его историю соревнований, среди участников когда-либо участвовавших в этом конкретном подразделении.

# 4.1 К полной таблице присоединим каждому участнику Общее количество его побед:

df3 = df.join(df2, on='name').filter(items=['name', 'division', 'sex', 'winnings'])
df3['winnings'] = df3['winnings'].fillna(0)
df3
```

```python
# 4.2 Найдем наибольшее Общее количество побед участника для каждой комбинации подразделения и пола:

df4 = pd.DataFrame(df3.groupby(['division', 'sex'])['winnings'].max().sort_values(ascending=False))
df4
```

```python
# 4.3 Добавим имена участников:

df5 = pd.merge(df3, df4, on=['division', 'sex', 'winnings']).drop_duplicates()\
                .sort_values(by=['winnings', 'division'], ascending=False).set_index('name')
df5

    # Таблицу получаем немного больше списка возможных комбинаций подразделения и пола, так как например:
        
        # division    sex    winnings
        # --------    ---    --------
        # Juniors      M       24.0
    
    # в этой комбинации есть два участника с одинаковым наибольшим Общим количеством побед.
```

```python
# 5.Подключитесь к базе данных employees и выполните с помощью pandas последний запрос из финального проекта по SQL. 
    # Убедитесь, что результаты загружены в pandas DataFrame.

import pymysql
import warnings
warnings.filterwarnings('ignore')

connector = pymysql.connect(host='127.0.0.1',port=int(3306),user="root",passwd='root',db='employees')

df6 = pd.read_sql_query('SELECT * FROM (SELECT *, SUM(month_salary) OVER (PARTITION BY dept_no ORDER BY dept_no,\
    month_salary, emp_no) AS cumulative_salary FROM (SELECT ede.dept_no, es.emp_no, ROUND(es.salary / 12, 2)\
    AS month_salary FROM employees.salaries AS es LEFT JOIN employees.dept_emp AS ede ON(ede.emp_no = es.emp_no\
    AND es.from_date BETWEEN ede.from_date AND ede.to_date) WHERE CURDATE() BETWEEN es.from_date AND es.to_date) AS s_1)\
    AS s_2 WHERE cumulative_salary < 500000;', connector)
df6
```
